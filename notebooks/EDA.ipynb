##### Import Libraries

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

##### Load Datasets

```python
# Load the Benin dataset 
benin_data = pd.read_csv('solar-measurements_benin-malanville_qc.csv', encoding='ISO-8859-1')

# Display the first few rows of the Benin dataset
print("Sample rows from Benin dataset:")
benin_data.head()
```

```python
# Load the Togo dataset 
togo_data = pd.read_csv('solar-measurements_togo-dapaong_qc.csv', encoding='ISO-8859-1')

# Display the first few rows of the Togo dataset
print("Sample rows from Togo dataset:")
togo_data.head()
```

```python
# Load the Sierra Leone dataset 
sierraleone_data = pd.read_csv('solar-measurements_sierraleone-bumbuna_qc.csv', encoding='ISO-8859-1')

# Display the first few rows of the Sierra Leone dataset
print("Sample rows from Sierra Leone dataset:")
sierraleone_data.head()
```

##### Summary Statistics

```python
# Define the numeric columns based on what's in the dataset
numeric_columns = [
    'GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'Tamb', 'RH', 'WS', 
    'WSgust', 'WSstdev', 'WD', 'WDstdev', 'BP', 'Precipitation', 
    'TModA', 'TModB'
]

# Convert these columns to numeric (in case they are not already)
benin_data[numeric_columns] = benin_data[numeric_columns].apply(pd.to_numeric, errors='coerce')
sierraleone_data[numeric_columns] = sierraleone_data[numeric_columns].apply(pd.to_numeric, errors='coerce')
togo_data[numeric_columns] = togo_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Summary statistics for these numeric columns
benin_summary = benin_data[numeric_columns].describe()
sierra_leone_summary = sierraleone_data[numeric_columns].describe()
togo_summary = togo_data[numeric_columns].describe()

# Display the summary statistics
print("Benin Dataset Summary:")
print(benin_summary)
print("\nSierra Leone Dataset Summary:")
print(sierra_leone_summary)
print("\nTogo Dataset Summary:")
print(togo_summary)
```

##### Data Quality Check

```python
# Checking for missing values
print("Missing Values in Benin Data:")
print(benin_data.isnull().sum())

print("\nMissing Values in Sierra Leone Data:")
print(sierraleone_data.isnull().sum())

print("\nMissing Values in Togo Data:")
print(togo_data.isnull().sum())
```

##### Data Cleaning

```python
# Replace negative values in GHI, DNI, and DHI columns with 0
ghi_cols = ['GHI', 'DNI', 'DHI']

benin_data[ghi_cols] = benin_data[ghi_cols].clip(lower=0)
sierraleone_data[ghi_cols] = sierraleone_data[ghi_cols].clip(lower=0)
togo_data[ghi_cols] = togo_data[ghi_cols].clip(lower=0)

print("\nNegative values replaced with 0 in GHI, DNI, and DHI columns.")
```

```python
# Define thresholds for identifying outliers
moda_threshold = 1000  # Example threshold for ModA and ModB
wind_speed_threshold = 50  # Example threshold for WS, WSgust

# Cap the outliers for ModA, ModB, WS, and WSgust
sensor_cols = ['ModA', 'ModB']
wind_cols = ['WS', 'WSgust']

benin_data[sensor_cols] = benin_data[sensor_cols].clip(upper=moda_threshold)
benin_data[wind_cols] = benin_data[wind_cols].clip(upper=wind_speed_threshold)

sierraleone_data[sensor_cols] = sierraleone_data[sensor_cols].clip(upper=moda_threshold)
sierraleone_data[wind_cols] = sierraleone_data[wind_cols].clip(upper=wind_speed_threshold)

togo_data[sensor_cols] = togo_data[sensor_cols].clip(upper=moda_threshold)
togo_data[wind_cols] = togo_data[wind_cols].clip(upper=wind_speed_threshold)

print("\nOutliers capped in sensor readings and wind speed data.")
```

##### Time Series Analysis

```python
# Drop rows where 'Timestamp' equals the placeholder
benin_data = benin_data[benin_data['Timestamp'] != 'yyyy-mm-dd hh:mm']
togo_data = togo_data[togo_data['Timestamp'] != 'yyyy-mm-dd hh:mm']
sierraleone_data = sierraleone_data[sierraleone_data['Timestamp'] != 'yyyy-mm-dd hh:mm']

# Convert the remaining 'Timestamp' column to datetime with the correct format
benin_data['Timestamp'] = pd.to_datetime(benin_data['Timestamp'], format="%Y-%m-%d %H:%M")
togo_data['Timestamp'] = pd.to_datetime(togo_data['Timestamp'], format="%Y-%m-%d %H:%M")
sierraleone_data['Timestamp'] = pd.to_datetime(sierraleone_data['Timestamp'], format="%Y-%m-%d %H:%M")

# Checking if the conversion was successful
print(benin_data['Timestamp'].head())
print(togo_data['Timestamp'].head())
print(sierraleone_data['Timestamp'].head())
```

```python
def plot_time_series(data, columns, title, colors):
    plt.figure(figsize=(14, 8))

    # Plot each column with its corresponding color
    for col, color in zip(columns, colors):
        plt.plot(data['Timestamp'], data[col], label=col, color=color)

    # Add title and labels
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Values')
    plt.legend()
    plt.grid(True)
    
    # Display the plot
    plt.show()

# Define the columns to plot and their corresponding colors
columns_to_plot = ['GHI', 'DNI', 'DHI', 'Tamb']
colors = ['orange', 'blue', 'green', 'red']

plot_time_series(benin_data, columns_to_plot, 'Time Series of GHI, DNI, DHI, and Tamb for Benin', colors)
plot_time_series(togo_data, columns_to_plot, 'Time Series of GHI, DNI, DHI, and Tamb for Togo', colors)
plot_time_series(sierraleone_data, columns_to_plot, 'Time Series of GHI, DNI, DHI, and Tamb for Sierra Leone', colors)
```

##### Cleaning impact

```python
def analyze_cleaning_impact(data, cleaning_column, sensor_columns, title_prefix):
    # Separate data before and after cleaning
    before_cleaning = data[data[cleaning_column] == 0]
    after_cleaning = data[data[cleaning_column] == 1]

    # Summary statistics for specified sensor columns before and after cleaning
    print("Before Cleaning:")
    print(before_cleaning[sensor_columns].describe(), "\n")

    print("After Cleaning:")
    print(after_cleaning[sensor_columns].describe(), "\n")

    # Plotting sensor columns over time
    plt.figure(figsize=(14, 6))

    for i, col in enumerate(sensor_columns):
        plt.subplot(len(sensor_columns), 1, i + 1)
        plt.plot(before_cleaning['Timestamp'], before_cleaning[col], label='Before Cleaning', color='red')
        plt.plot(after_cleaning['Timestamp'], after_cleaning[col], label='After Cleaning', color='green')
        plt.title(f'{title_prefix} Impact on {col}')
        plt.xlabel('Time')
        plt.ylabel(col)
        plt.legend()

    plt.tight_layout()
    plt.show()

# Define the columns to analyze and the title prefix
sensor_columns = ['ModA', 'ModB']
title_prefix = 'Impact of Cleaning on'

# Analyze and plot the impact of cleaning on ModA and ModB
analyze_cleaning_impact(benin_data, 'Cleaning', sensor_columns, title_prefix)
analyze_cleaning_impact(togo_data, 'Cleaning', sensor_columns, title_prefix)
analyze_cleaning_impact(sierraleone_data, 'Cleaning', sensor_columns, title_prefix)

```

##### Correlation Analysis

```python
def plot_correlation_heatmap(df):
    corr = df[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']].corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap')
    plt.show()


plot_correlation_heatmap(benin_data)
plot_correlation_heatmap(sierraleone_data)
plot_correlation_heatmap(togo_data)
```

##### Wind Analysis

```python
def plot_wind_polar(df):
    plt.figure(figsize=(8, 8))
    plt.subplot(projection='polar')
    theta = np.deg2rad(df['WD'].dropna())
    r = df['WS'].dropna()
    plt.scatter(theta, r, c='b', alpha=0.5)
    plt.title('Wind Speed Distribution')
    plt.show()


plot_wind_polar(benin_data)
plot_wind_polar(sierraleone_data)
plot_wind_polar(togo_data)
```

##### Histograms

```python
def plot_histogram(df, column):
    plt.figure(figsize=(10, 6))
    plt.hist(df[column].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()

# Example: Histogram of GHI
plot_histogram(benin_data, 'GHI')
plot_histogram(sierraleone_data, 'GHI')
plot_histogram(togo_data, 'GHI')
```

##### Z-Score Analysis

```python
def calculate_z_scores(dataframe, columns, threshold=3):
    flagged_df = dataframe.copy()
    flagged_summary = {}

    for column in columns:
        # Calculate the mean and standard deviation for the column
        mean = flagged_df[column].mean()
        std = flagged_df[column].std()
        
        # Calculate the Z-scores
        flagged_df[f'Z_{column}'] = (flagged_df[column] - mean) / std
        
        # Flag data points that exceed the threshold
        flagged_df[f'Flagged_{column}'] = flagged_df[f'Z_{column}'].apply(lambda z: abs(z) > threshold)
        
        # Count the number of flagged data points
        flagged_count = flagged_df[f'Flagged_{column}'].sum()
        flagged_summary[column] = flagged_count

    # Convert flagged summary to DataFrame
    flagged_summary_df = pd.DataFrame(flagged_summary, index=['Flagged_Count']).T
    
    return flagged_df, flagged_summary_df

columns_for_z_score_analysis = ['GHI', 'DNI', 'DHI', 'WS', 'TModA', 'TModB']
flagged_data, flagged_summary = calculate_z_scores(benin_data, columns_for_z_score_analysis)

# Display flagged data points summary for inspection
print(flagged_summary)
```

```python
columns_for_z_score_analysis = ['GHI', 'DNI', 'DHI', 'WS', 'TModA', 'TModB']
flagged_data, flagged_summary = calculate_z_scores(togo_data, columns_for_z_score_analysis)

# Display flagged data points summary for inspection
print(flagged_summary)
```

```python
columns_for_z_score_analysis = ['GHI', 'DNI', 'DHI', 'WS', 'TModA', 'TModB']
flagged_data, flagged_summary = calculate_z_scores(sierraleone_data, columns_for_z_score_analysis)

# Display flagged data points summary for inspection
print(flagged_summary)
```

##### Bubble charts

```python
def plot_bubble_chart(data, x_var, y_var, bubble_size_var, bubble_color_var, title):
    plt.figure(figsize=(10, 6))

    # Normalize bubble sizes for better visualization
    sizes = data[bubble_size_var] / data[bubble_size_var].max() * 1000

    # Create bubble chart
    scatter = plt.scatter(data[x_var], data[y_var], s=sizes, c=data[bubble_color_var], cmap='viridis', alpha=0.6, edgecolors="w", linewidth=0.5)

    plt.title(title)
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    
    # Add a color bar to show the mapping of bubble color variable
    cbar = plt.colorbar(scatter)
    cbar.set_label(bubble_color_var)

    plt.grid(True)
    plt.show()

plot_bubble_chart(
    benin_data, 
    x_var='GHI', 
    y_var='Tamb', 
    bubble_size_var='RH', 
    bubble_color_var='BP', 
    title='Bubble Chart: GHI vs. Tamb with Bubble Size by RH and Color by BP'
)

plot_bubble_chart(
    togo_data, 
    x_var='GHI', 
    y_var='Tamb', 
    bubble_size_var='RH', 
    bubble_color_var='BP', 
    title='Bubble Chart: GHI vs. Tamb with Bubble Size by RH and Color by BP'
)

plot_bubble_chart(
    sierraleone_data, 
    x_var='GHI', 
    y_var='Tamb', 
    bubble_size_var='RH', 
    bubble_color_var='BP', 
    title='Bubble Chart: GHI vs. Tamb with Bubble Size by RH and Color by BP'
)
```
